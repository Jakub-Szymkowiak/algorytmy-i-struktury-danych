```python
class BST:
    class Node:
        def __init__(self, key: int):
            # klucz przechowywany w węźle
            self.key = key

            # wskaźnik na lewe poddrzewo (wszystkie klucze < key)
            self.left = None

            # wskaźnik na prawe poddrzewo (wszystkie klucze > key)
            self.right = None

    def __init__(self):
        # korzeń drzewa (na początku drzewo jest puste)
        self.root = None

    def is_empty(self) -> bool:
        # zwraca True, jeśli drzewo jest puste
        return self.root is None

    def search(self, x: int) -> bool:
        # sprawdza, czy w drzewie znajduje się klucz x

        # zaczynamy w korzeniu
        cur = self.root

        # dopóki nie spadniemy poza drzewo
        while cur is not None:

            # jeśli trafiliśmy w x, to znaleźliśmy element
            if x == cur.key:
                return True

            # jeśli x jest mniejsze, idziemy do lewego poddrzewa
            if x < cur.key:
                cur = cur.left

            # jeśli x jest większe, idziemy do prawego poddrzewa
            else:
                cur = cur.right

        # jeśli zeszliśmy do None, to elementu nie ma w drzewie
        return False

    def insert(self, x: int) -> None:
        # wstawia klucz x do drzewa (bez duplikatów)

        # jeśli drzewo jest puste, x staje się korzeniem
        if self.root is None:
            self.root = self.Node(x)
            return

        # zaczynamy w korzeniu
        cur = self.root

        # schodzimy w dół, aż znajdziemy miejsce na nowy węzeł
        while True:

            # jeśli x już istnieje, to nic nie robimy (ignorujemy duplikaty)
            if x == cur.key:
                return

            # jeśli x jest mniejsze, próbujemy wejść w lewo
            if x < cur.key:

                # jeśli lewego dziecka nie ma, to tutaj wstawiamy nowy węzeł
                if cur.left is None:
                    cur.left = self.Node(x)
                    return

                # jeśli lewe dziecko istnieje, schodzimy dalej
                cur = cur.left

            # jeśli x jest większe, próbujemy wejść w prawo
            else:

                # jeśli prawego dziecka nie ma, to tutaj wstawiamy nowy węzeł
                if cur.right is None:
                    cur.right = self.Node(x)
                    return

                # jeśli prawe dziecko istnieje, schodzimy dalej
                cur = cur.right

    def inorder(self) -> list[int]:
        # zwraca listę kluczy w kolejności inorder (lewo -> węzeł -> prawo)
        # dla BST daje to klucze w porządku rosnącym

        # lista wynikowa
        out = []

        # stos do symulowania rekurencji
        stack = []

        # zaczynamy od korzenia
        cur = self.root

        # dopóki mamy gdzie iść w dół lub mamy coś na stosie
        while cur is not None or stack:

            # schodzimy maksymalnie w lewo, odkładając po drodze węzły na stos
            while cur is not None:
                stack.append(cur)
                cur = cur.left

            # bierzemy węzeł ze stosu (to jest "następny" w inorder)
            cur = stack.pop()

            # dopisujemy jego klucz do wyniku
            out.append(cur.key)

            # przechodzimy do prawego poddrzewa
            cur = cur.right

        # zwracamy listę kluczy
        return out
```
