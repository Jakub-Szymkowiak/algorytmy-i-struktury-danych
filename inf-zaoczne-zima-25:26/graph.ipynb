### Przeszukiwanie grafu

```python
# Nasza implementacja grafu

class Graph:
    def __init__(self, n: int, directed: bool=False):
        # liczba wierzchołków
        self.n = n

        # czy graf jest skierowany
        self.directed = directed

        # lista sąsiedztwa – zainicjalizowana jako pusta
        self.adj = [[] for _ in range(n)]

    def add_edge(self, u: int, v: int):
        # dodaje krawędź między wierzchołkami u i v
        self.adj[u].append(v)

        # jeśli graf nie jest skierowany
        if not self.directed:
            # dodaje krawędź w drugą stronę
            self.adj[v].append(u)

    def neighbors(self, u: int):
        # zwraca listę sąsiadów wierzchołka u
        return self.adj[u]
```

```python
# Przykład użycia:
g = Graph(7)

g.add_edge(1, 2)
g.add_edge(1, 5)
g.add_edge(1, 6)
g.add_edge(2, 5)
g.add_edge(2, 3)
g.add_edge(3, 4)
g.add_edge(4, 5)

print(g.neighbors(1))  # powinno wypisać [0, 3, 4]
print(g.adj)           # wypisuje całą listę sąsiedztwa grafu

3 in g.neighbors(1)  # sprawdza, czy istnieje krawędź 1 -> 3
```

#### Struktury pomocnicze

```python
# Do BFS wykorzystamy kolejkę

from collections import deque

class Queue:
    def __init__(self):
        self._d = deque()

    def enqueue(self, x):
        self._d.append(x)

    def dequeue(self):
        return self._d.popleft()

    def is_empty(self) -> bool:
        return not self._d
    
    
# Do DFS wykorzystamy stos

class Stack:
    def __init__(self):
        self._a = []

    def push(self, x):
        self._a.append(x)

    def pop(self):
        return self._a.pop()

    def is_empty(self) -> bool:
        return len(self._a) == 0
```

#### Przeszukiwanie wszerz

```python
def bfs(g: Graph, start: int):
    # g – graf do przeszukania
    # start – wierzchołek startowy

    # tablica odwiedzonych wierzchołków
    # na początku wszystkie wierzchołki są nieodwiedzone
    visited = [False] * g.n

    # odległości od wierzchołka startowego (liczba krawędzi)
    dist = [-1] * g.n

    # poprzednik na najkrótszej ścieżce
    parent = [-1] * g.n

    # kolejka - zaczynamy od wierzchołka startowego
    q = Queue()
    q.enqueue(start)

    # oznaczamy wierzchołek startowy jako odwiedzony
    visited[start] = True

    # odległość od wierzchołka startowego do samego siebie wynosi 0
    dist[start] = 0

    # tak długo, jak są wierzchołki w kolejce
    while not q.is_empty():

        # zdejmujemy wierzchołek z kolejki
        u = q.dequeue()

        # dla każdego sąsiada v wierzchołka u
        for v in g.neighbors(u):
            # jeśli wierzchołek v nie był jeszcze odwiedzony
            if not visited[v]:
                # oznaczamy go jako odwiedzony
                visited[v] = True
                # ustawiamy jego odległość
                dist[v] = dist[u] + 1
                # ustawiamy jego poprzednika
                parent[v] = u
                # dodajemy go do kolejki
                q.enqueue(v)

    return dist, parent
```

```python
d, p = bfs(g, 1)
print(d, p)

# d[v] – ile krawędzi ma najkrótsza ścieżka od wierzchołka 1 do wierzchołka v

# p[v] – z którego wierzchołka po raz pierwszy odkryliśmy wierzchołek v

tree_edges = [(p[v], v) for v in range(g.n) if d[v] != -1]
print(tree_edges)  # wypisuje krawędzie drzewa BFS
```

```python
def shortest_path(g: Graph, start, target):
    # najkrótsza ścieżka w grafie nieważonym (BFS)
    # zwraca listę wierzchołków od start do target; [] jeśli nieosiągalny

    dist, parent = bfs(g, start)

    # jeśli target nieosiągalny
    if dist[target] == -1:
        return []

    # odtwarzamy ścieżkę cofając się po rodzicach
    path = []
    v = target
    while v != start:
        path.append(v)
        v = parent[v]
    path.append(start)
    path.reverse()

    # odwracamy ścieżkę, bo zaczęliśmy od target
    path.reverse()
    return path
```

#### Przeszukiwanie wgłąb – praca samodzielna

```python

def dfs(g: Graph, start: int):
    # g – graf do przeszukania
    # start – wierzchołek startowy

    # tablica odwiedzonych wierzchołków
    # na początku wszystkie wierzchołki są nieodwiedzone
    visited = [False] * g.n

    # kolejność odwiedzania wierzchołków
    order = []

    # stos - zaczynamy od wierzchołka startowego
    stack = Stack()
    stack.push(start)

    # oznaczamy wierzchołek startowy jako odwiedzony
    visited[start] = True

    # tak długo, jak są wierzchołki na stosie
    while not stack.is_empty():

        # zdejmujemy wierzchołek ze stosu
        u = stack.pop()

        # zapisujemy kolejność odwiedzania
        order.append(u)

        # dla każdego sąsiada v wierzchołka u
        for v in g.neighbors(u):
            # jeśli wierzchołek v nie był jeszcze odwiedzony
            if not visited[v]:
                # oznaczamy go jako odwiedzony
                visited[v] = True
                # dodajemy go na stos
                stack.push(v)

    return order
```
