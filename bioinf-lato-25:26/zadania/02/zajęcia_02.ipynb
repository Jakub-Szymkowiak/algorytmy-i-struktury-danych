##### Zadanie 2

Niech `A: list[int]` będzie tablicą liczb całkowitych o długości `n > 0`. 
1. Jak znaleźć największą liczbę spośród liczb w tablicy `A`?
2. Znajdź indeks największej liczby w tablicy. Jeśli w tablicy jest kilka liczb o tej samej największej wartości, znajdź indeks pierwszej z nich.

```python
A = [7, -3, 12, 5, 12, 1]

assert len(A) > 0, "Lista nie może być pusta"

# wariant 1
_max = A[0]
for i in range(1, len(A)):
    if A[i] > _max:
        _max = A[i]

print(_max)

# wariant 2
argmax = 0
for i in range(1, len(A)):
    if A[i] > A[argmax]:
        argmax = i

print(argmax)
print(A[argmax])
```

##### Zadanie 3

Niech `A: list[int]` będzie tablicą liczb całkowitych o długości `n > 0`.

- Znajdź liczbę liczb dodatnich (liczbę indeksów `i`, dla których `A[i] > 0`).
- Znajdź ich sumę i średnią.

Niech ponadto `k: int` będzie liczbą całkowitą dodatnią.

- Znajdź liczbę wystąpień liczby `k` w tablicy `A`.
- Znajdź liczbę liczb w tablicy `A` podzielnych przez `k`.

```python
A = [4, -2, 7, 0, 7, 12, -5, 14]
k = 7

assert len(A) > 0, "Lista nie może być pusta"
assert k > 0, "k musi być dodatnie"

# wariant 1: liczby dodatnie
positive_count = 0
positive_sum = 0

for value in A:
    if value > 0:
        positive_count += 1
        positive_sum += value

if positive_count > 0:
    positive_avg = positive_sum / positive_count
else:
    positive_avg = None

print(positive_count)
print(positive_sum)
print(positive_avg)

# wariant 2: liczba wystąpień k oraz liczba elementów podzielnych przez k
count_equal_k = 0
count_divisible_by_k = 0

for value in A:
    if value == k:
        count_equal_k += 1
    if value % k == 0:
        count_divisible_by_k += 1

print(count_equal_k)
print(count_divisible_by_k)
```

##### Zadanie 4

Niech `A: list[int]` będzie tablicą liczb całkowitych o długości `n > 0`. Znajdź tablicę `B: list[int]`, która będzie zawierać dokładnie te same wartości, co `A`, ale w odwrotnej kolejności.

```python
A = [3, 1, 4, 1, 5, 9]

assert len(A) > 0, "Lista nie może być pusta"

B = []
for i in range(len(A) - 1, -1, -1):
    B.append(A[i])

print(A)
print(B)
```

##### Zadanie 5

Dana jest liczba `x: float`, stopień wielomianu `n > 0` oraz tablica `W: list[float]` długości `n+1` zawierająca współczynniki wielomianu:

`f(x) = W[0] + W[1]x + W[2]x^2 + ... + W[n-1]x^(n-1) + W[n]x^n`

(czyli: `W[m]` jest współczynnikiem przy `m`-tej potędze). Znajdź wartość wielomianu w punkcie `x` (oblicz `f(x)`).

```python
x = 2.0
W = [1.5, -3.0, 0.5, 4.0]  # 1.5 - 3x + 0.5x^2 + 4x^3

assert len(W) >= 2, "Tablica współczynników musi mieć długość co najmniej 2"

value = 0.0
for power in range(len(W)):
    value += W[power] * (x ** power)

print(value)
```

##### Zadanie 6

Zauważ, że wielomian `f(x)` możemy zapisać również w następujący sposób:

`f(x) = W[0] + x(W[1] + x(W[2] + ... + x(W[n-1] + xW[n])...))`

Korzystając z tego faktu, znajdź wartość wielomianu w punkcie `x` bez obliczania kolejnych potęg `x`.

```python
x = 2.0
W = [1.5, -3.0, 0.5, 4.0]

assert len(W) >= 2, "Tablica współczynników musi mieć długość co najmniej 2"

# wariant WHILE
i = len(W) - 1
value = W[i]

i -= 1
while i >= 0:
    value = W[i] + x * value
    i -= 1

print(value)

# wariant FOR (krótszy, ale brzydszy)
value = W[-1]
for i in range(len(W) - 2, -1, -1):
    value = W[i] + x * value

print(value)
```
