```python
class MinHeap:
    def __init__(self):
        # kopiec trzymamy w tablicy
        self.a = []

    def is_empty(self) -> bool:
        # zwraca True, jeśli kopiec jest pusty
        return len(self.a) == 0

    def peek(self):
        # zwraca najmniejszy element (korzeń) bez usuwania go z kopca

        # jeśli kopiec jest pusty, nie ma czego zwrócić
        if not self.a or self.is_empty():
            raise IndexError("Kopiec jest pusty!")

        # w min-heapie najmniejszy element zawsze jest pod indeksem 0
        return self.a[0]

    def push(self, x):
        # dodaje element x do kopca

        # dodajemy element na koniec tablicy
        # (czyli na najniższy wolny liść w drzewie kopca)
        self.a.append(x)

        # indeks wstawionego elementu
        i = len(self.a) - 1

        # podciągamy element w górę, dopóki własność kopca jest złamana
        # (dla min-heapa: rodzic ma być <= dziecka)
        while i > 0:
            # indeks rodzica elementu i
            p = (i - 1) // 2

            # jeśli rodzic <= dziecko, własność kopca jest spełniona
            if self.a[p] <= self.a[i]:
                break

            # w przeciwnym razie zamieniamy element z rodzicem
            self.a[p], self.a[i] = self.a[i], self.a[p]

            # przechodzimy do rodzica i kontynuujemy podciąganie
            i = p

    def pop(self):
        # usuwa i zwraca najmniejszy element (korzeń) z kopca

        # jeśli kopiec jest pusty, nie ma czego usunąć
        if not self.a or self.is_empty():
            raise IndexError("Kopiec jest pusty!")

        # zapamiętujemy najmniejszy element (korzeń)
        top = self.a[0]

        # zdejmujemy ostatni element z tablicy
        last = self.a.pop()

        # jeśli po zdjęciu ostatniego elementu kopiec nie jest pusty
        if self.a:
            # przenosimy ostatni element na górę (do korzenia)
            self.a[0] = last

            # spychamy go w dół, żeby przywrócić własność kopca
            self._sift_down(0)

        # zwracamy element, który był na górze kopca
        return top

    def _sift_down(self, i: int):
        # spycha element o indeksie i w dół, aż własność kopca będzie spełniona

        # aktualny rozmiar kopca
        n = len(self.a)

        # powtarzamy, dopóki możemy zejść niżej
        while True:
            # indeks lewego dziecka
            l = 2 * i + 1

            # indeks prawego dziecka
            r = 2 * i + 2

            # zakładamy, że najmniejszy jest aktualny węzeł
            smallest = i

            # jeśli lewe dziecko istnieje i jest mniejsze od aktualnego minimum
            if l < n and self.a[l] < self.a[smallest]:
                smallest = l

            # jeśli prawe dziecko istnieje i jest mniejsze od aktualnego minimum
            if r < n and self.a[r] < self.a[smallest]:
                smallest = r

            # jeśli aktualny węzeł jest już najmniejszy, to kończymy
            if smallest == i:
                break

            # w przeciwnym razie zamieniamy go z mniejszym dzieckiem
            self.a[i], self.a[smallest] = self.a[smallest], self.a[i]

            # schodzimy do pozycji, na którą spadł element
            i = smallest
```

```python
h = MinHeap()
for x in [5, 2, 8, 1, 7, 3]:
    h.push(x)

print(h.peek())     # 1
print(h.pop())      # 1
print(h.pop())      # 2
print(h.pop())      # 3
print(h.a)          # reszta w strukturze kopca (nieposortowana)
```

```python
from dataclasses import dataclass
from typing import Any

# Klasa pomocnicza: element kolejki priorytetowej, który przechowuje zarówno priorytet, jak i ładunek (payload)
@dataclass
class PQItem:
    priority: int
    payload: Any

    def __lt__(self, other: "PQItem") -> bool:
        return self.priority < other.priority

    def __le__(self, other: "PQItem") -> bool:
        return self.priority <= other.priority

# Przykład działania klasy PQItem
print(PQItem(1, "zadanie A").priority)  # 1
print(PQItem(1, "zadanie A").payload)   # "zadanie A"
print(PQItem(1, "zadanie A") < PQItem(2, "zadanie B"))  # True     
```
