W poniższej komórce znajduje się implementacja klasy `BinarySearchTree` wyposażona w dwie metody:
1. `.insert(value)` – pozwala na umieszczenie danych liczbowych w drzewie
2. `.__str__()` – formatuje wynik komendy `print` zastosowanej na obiekcie klasy `BinarySearchTree` 

````python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = Node(value)
            else:
                self._insert(node.left, value)
        else:
            if node.right is None:
                node.right = Node(value)
            else:
                self._insert(node.right, value)

    def __str__(self):
        if self.root is None:
            return "∅ Empty BST"

        lines = []

        def draw(node, prefix="", is_tail=True):
            if node.right is not None:
                draw(
                    node.right,
                    prefix + ("│   " if is_tail else "    "),
                    False,
                )

            connector = "└── " if is_tail else "┌── "
            lines.append(prefix + connector + str(node.value))

            if node.left is not None:
                draw(
                    node.left,
                    prefix + ("    " if is_tail else "│   "),
                    True,
                )

        lines.append("BST")
        draw(self.root)
        return "\n".join(lines)
````

````python
### Przykład

# Tworzymy nową instancję BST
bst = BinarySearchTree()

# Wstawiamy przykładowe wartości
for value in [10, 5, 15, 3, 7, 12, 20, 100, 3, 1, 51, 45, 21, 14, 13, 17, 16, 18, 19, 25, 30, 28, 27, 26, 29]:
    bst.insert(value)

# Wyświetlamy strukturę drzewa
print(bst)
````

### Przechodzenie przez wszystkie węzły BST (_tree traversal_)

Chcemy odwiedzić wszystkie węzły BST. 
Są na to trzy standardowe sposoby, które wykorzystują rekurencję:

- *inorder* – `left -> node -> right`: 
    1. Wywołujemy procedurę dla lewego poddrzewa.
    2. Wyświetlamy węzeł startowy.
    3. Wywołujemy procedurę dla prawego poddrzewa.

- *preorder* – `node -> left -> right`:
    1. Wyświetlamy węzeł startowy.
    2. Wywołujemy procedurę dla lewego poddrzewa.
    3. Wywołujemy procedurę dla prawego poddrzewa.

- *postorder* – `left -> right -> node`:
    1. Wywołujemy procedurę dla lewego poddrzewa.
    2. Wywołujemy procedurę dla prawego poddrzewa.
    3. Wyświetlamy węzeł startowy.

Na przykład, dla BST postaci:
```
        10
       /  \
      5    15
     / \
    3   7
```
otrzymamy:

- inorder: `3, 5, 7, 10, 15`
- preorder: `10, 5, 3, 7, 15`
- postorder: `3, 7, 5, 15, 10`
